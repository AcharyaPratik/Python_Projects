```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import csv file
df = pd.read_csv('Student_Stress_Dataset.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
df.columns
```

```python
#rename column
df = df.rename(columns= {'Student_ID':'ID'})
```

```python
# Replace 'F' with 'Female' and 'M' with 'Male'
df['Gender'] = df['Gender'].replace('F', 'Female')
df['Gender'] = df['Gender'].replace('M', 'Male')
df
```

```python
# Replace '0' with 'No' and '1' with 'Yes'
df['Counseling_Need'] = df['Counseling_Need'].replace(0, 'No')
df['Counseling_Need'] = df['Counseling_Need'].replace(1, 'Yes')
df['Risky_Behaviors_Engaged'] = df['Risky_Behaviors_Engaged'].replace(0, 'No')
df['Risky_Behaviors_Engaged'] = df['Risky_Behaviors_Engaged'].replace(1, 'Yes')
df
```

```python
#describe() method returns description of the data in the DataFrame
df.describe()
```

```python
#we can use describe() for specific columns as well
df[['Age', 'Sleep_Hours', 'Total_Stress_Level']].describe()
```

# Exploratory Data Analysis

```python
#Gender Distribution
plt.figure(figsize=(5,4))
a = sns.countplot(data = df, x = 'Gender')
a.bar_label(a.containers[0])
plt.title('Gender Distribution')
plt.show()
```

##### We can see that male have higher number count than the female

```python
#Gender VS CGPA
plt.figure(figsize=(6, 5)) 
sns.boxplot(x='Gender', y='CGPA', data=df) 
plt.title('Gender VS CGPA') 
plt.xlabel('Gender')
plt.ylabel('CGPA') 
plt.grid(True) 
plt.show()
```

##### We conclude that the male have a little higher average CGPA than the female

```python
#Age Distribution
age_counts = df['Age'].value_counts().sort_index()
b = age_counts.plot(kind='bar')
b.bar_label(b.containers[0])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count of students')
plt.show()
```

##### This give the distribution of age groups i.e. there are most students of age 24 and least students of age 18

```python
#Age distribution for different age groups
c = sns.countplot(data = df, x = 'Age', hue = 'Gender')
for bars in c.containers:
    c.bar_label(bars)
plt.ylabel('Count of students')
plt.show()
```

##### This give more compact distribution of age groups among gender

```python
#Total level of stress different age groups
plt.figure(figsize=(7, 4)) 
sns.lineplot(x='Age', y='Total_Stress_Level', data=df, marker='o') 
plt.title('Total Stress Level by Age') 
plt.xlabel('Age') 
plt.ylabel('Total Stress Level') 
plt.grid(True) 
plt.show()
```

##### In the chart we can see that the student of age group 19 and 21 have the highest stress level

```python
#Distribution of Stress Sources Across Different Ages
plt.figure(figsize=(10, 6)) 
e = sns.countplot(x='Age', hue='Main_Stress_Source', data=df) 
plt.title('Distribution of Stress Sources Across Different Ages') 
plt.xlabel('Age') 
plt.ylabel('Count of students') 
plt.legend(title='Stress Source') 
for bars in e.containers:
    e.bar_label(bars)
plt.show()
```

##### We can see the different level of  strees sources across different ages

```python
#Age VS CGPA 
plt.figure(figsize=(7, 4)) 
avg_cgpa = df.groupby('Age')['CGPA'].mean().reset_index()
sns.lineplot(x='Age', y='CGPA', data=avg_cgpa, marker='o')   
plt.title('Average CGPA vs. Age') 
plt.xlabel('Age') 
plt.ylabel('CGPA') 
plt.grid(True)
plt.show()
```

##### This chart shows the average CGPA of age group 23 is the highest and that of 19 is the lowest

```python
#Students distribution across different year of study
age_counts = df['Year_of_Study'].value_counts().sort_index()
b = age_counts.plot(kind='bar')
b.bar_label(b.containers[0])
plt.title('Students distribution across different year of study')
plt.xlabel('Year of students')
plt.ylabel('Count of students')
plt.show()
```

##### There are more students in 2nd year

```python
#Year of Study VS Total Stress Level 
plt.figure(figsize=(7, 4)) 
avg_stress = df.groupby('Year_of_Study')['Total_Stress_Level'].mean().reset_index()
sns.lineplot(x='Year_of_Study', y='Total_Stress_Level', data=avg_stress, marker='o') 
plt.title('Total Stress Level by Year of Study') 
plt.xlabel('Year of Study') 
plt.ylabel('Total Stress Level') 
plt.grid(True) 
plt.show()
```

##### Students of 3rd year have more stress levels than students of other years can be concluded by the above chart

```python
#CGPA acquired by each year's students 
plt.figure(figsize=(7, 4)) 
i = sns.boxplot(x='Year_of_Study', y='CGPA', data=df) 
plt.title('CGPA by Year of Study') 
plt.xlabel('Year of Study') 
plt.ylabel('CGPA') 
plt.grid(True) 
plt.show()
```

##### This chart shows that the 2nd year's students scores more avereage CGPA compared to others.

```python
#Distribution of Student Behavior by Family Support
plt.figure(figsize=(10, 6)) 
ax = sns.countplot(x='Family_Support', hue='Student_Behavior', data=df) 
plt.title('Distribution of Student Behavior by Family Support') 
plt.xlabel('Family Support Level') 
plt.ylabel('Count of students') 
plt.legend(title='Student Behavior')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

##### This chart shows that the behavior of students based on their family support

```python
#Average CGPA of students having different levels of family support
plt.figure(figsize=(7, 4)) 
avg_cgpa = df.groupby('Family_Support')['CGPA'].mean().reset_index()
sns.lineplot(x='Family_Support', y='CGPA', data=avg_cgpa, marker='o')
plt.title('Average CGPA vs. Family Support') 
plt.xlabel('Age') 
plt.ylabel('CGPA') 
plt.grid(True)
plt.show()
```

##### The children of family giving the medium family support are scoring more CGPA on average

```python
#Engaged on Risky Behaviors on the basis of total Stress Level 
plt.figure(figsize=(5, 4)) 
f = sns.barplot(x='Risky_Behaviors_Engaged', y='Total_Stress_Level', data=df) 
plt.title('Impact of Risky Behaviors on Total Stress Level') 
plt.xlabel('Risky Behaviors Engaged') 
plt.ylabel('Total Stress Level') 
f.bar_label(f.containers[0])
plt.show()
```

##### There is not much difference in the students being engaged in risky behaviors on the basis of total Stress Level

```python
#Sleep Hours VS Average CGPA 
plt.figure(figsize=(7, 4)) 
avg_cgpa = df.groupby('Sleep_Hours')['CGPA'].mean().reset_index()
sns.lineplot(x='Sleep_Hours', y='CGPA', data=avg_cgpa, marker='o', color='RED')
plt.title('Relationship Between Sleep Hours and CGPA') 
plt.xlabel('Sleep Hours') 
plt.ylabel('CGPA') 
plt.grid(True) 
plt.show()
```

##### Students having sleep hours between 5 hours and 6 hours have a higher average CGPA

```python
plt.figure(figsize=(10, 5)) 
sns.boxplot(x='Academic_Stress_Level', y='CGPA', data=df) 
plt.title('Impact of Academic Stress on CGPA') 
plt.xlabel('Academic Stress Level')
plt.ylabel('CGPA') 
plt.grid(True) 
plt.show()
```

##### This boxplot shows the impact of Academic stress on the CGPA of students

```python
plt.figure(figsize=(10, 5)) 
sns.boxplot(x='Family_Stress_Level', y='CGPA', data=df) 
plt.title('Impact of Financial Stress on CGPA') 
plt.xlabel('Financial Stress Level')
plt.ylabel('CGPA') 
plt.grid(True) 
plt.show()
```

##### This boxplot shows the impact of financial stress on the CGPA of students

```python
#Students distribution across different year of study
coping = df["Coping_Mechanisms_Used"].value_counts()
b = coping.plot(kind='bar')
b.bar_label(b.containers[0])
plt.title('Coping mechanism distribution')
plt.xlabel('Coping Mechanisms')
plt.ylabel('Count of students')
plt.show()
```
